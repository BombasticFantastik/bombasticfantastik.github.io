```python
import os
start_idx=2
max_idx=62
```

```python
path='/home/artemybombastic/MyGit/ArtemyBombastic.ru/cursors/'

all_files=[file_name for file_name in os.listdir(path) if (('_' not in file_name) and ('.conf' not in file_name) and ('.ipynb' not in file_name  )) ] 
```

```python
for file in os.listdir(path):
    if file not in all_files and file!='.ipynb_checkpoints':
        os.remove(os.path.join(path,file))
```

```python
for file in all_files:
    os.system(f"xcur2png {file}")
    goal_indexes=[i for i in range(start_idx,max_idx,3)]
    for i in range(0,max_idx):
        if i not in goal_indexes:
            if os.path.join(f'{file}_00{i}.png') in os.listdir():
                print(i)
                os.remove(os.path.join(path,f'{file}_00{i}.png'))
            else:
                os.remove(os.path.join(path,f'{file}_0{i}.png'))
                
```
